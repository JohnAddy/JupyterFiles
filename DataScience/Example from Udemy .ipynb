```python
#Import pandas, numpy 
#Import the cufflinks library
#Import modules to work with data visualization offline
#Connect the Javascript to the notebook with the init_notebook_mode()method
#with the connected argument set to True
#Update to use cufflinks offline 

import pandas as pd 
import numpy as np
from plotly import __version__
%matplotlib inline

import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
import chart_studio.plotly as py
```

```python
#Use pandas to create a randomly generated Datafram 1 
#Use rand() method to generate a normally distributed dataframe with 100 rows 
#and 4 columns, use columns with .split() method to create list with columns A,B,C and D
#Use the .head()method to display the top 10 rows of data

df= pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
df.head(10)
```

```python
#Use pandas to create a dictionary generated DataFrame 2

df2=pd.DataFrame({'Category': ['A','B','C'], 'Values': [32,43,54]})
df2.head(10)
```

```python
# Create a new dataframe
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z':[5,4,3,2,1]})
df3.head()
```

```python
#Line plot allows us to click on the elements in the legend to hide 
#and display context which is pretty neat. 

df.iplot()
```

```python
#Scatter plots method with arguments kind which determined the plot type ('scatter'), x (x-axis variable), y (y-axis variable)
#and mode argument removes the line connections setup by default with plotly

df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
#Use the iplot() method with arguments, kind argument determines the plot type 
#kind='bar' which uses an ggregate method to group data = sum() method

df.sum().iplot(kind='bar')
```

```python
#Use the iplot() method with arguments, kind argument determines the plot type 
#kind='box' which uses an ggregate method to group data = sum() method

df.iplot(kind='box')
```

```python
#Use the .iplot() method with arguments kind (surface), and 
#colorscale to alter the plot color.

df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
# Use the .iplot() method with arguments kind (histogram), 
#and bins to specify the quantity of distribution buckets.

df['A'].iplot(kind='histogram', bins=40)
```

```python
#Use the .iplot() method with argument kind (spread lot ) 
#and conduct a conditional selection for the desired columns. 
#the spread plot also comes with line plot 

df[['A', 'B']].iplot(kind='spread')
```

```python
#Use the .iplot() method with arguments kind (define plot type),
#x (x-axis variable), 
#y (y-axis variable) and size to reference the data point sizes.
#size argument references the size of the data points

df.iplot(kind='bubble', x='A', y='B', size= 'C')
```
